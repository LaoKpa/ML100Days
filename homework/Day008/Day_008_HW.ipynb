```python
def plot_category_chart(df, feature, rotation=90, ascending=False, title=None, x_name=None, figsize=(10, 4)):
    """
    Parameters:
    ----------
    df: DataFrame
    feature: str
    rotation: int
    ascending: bool
    title: str
    x_name: list
    figsize: tuple
    
    Returns:
    -------
    matplotlib chart
    """
    plt.figure(figsize=figsize)
    if ascending == True:
        chart = sns.countplot(df[feature], order=df[feature].value_counts().index)
    else: 
        chart = sns.countplot(df[feature])
    chart.set_xticklabels(chart.get_xticklabels(), rotation=rotation, ha="right")
    plt.grid(axis="y")
    plt.title(title)
    if x_name is not None:
        plt.xticks(np.arange(len(x_name)), x_name)
    plt.tight_layout()
    plt.show()
```

```python
# TARGET value 0 means loan is repayed, value 1 means loan is not repayed.
plot_category_chart(df=app_train, feature="TARGET", x_name=["Repayed", "Not Repayed"], rotation=0)
```

```python
plot_category_chart(df=app_train, feature="NAME_CONTRACT_TYPE", rotation=0)
```

```python
# Remove rows which CODE_GENDER is "XNA"
app_train_gender = app_train.loc[app_train['CODE_GENDER'] != "XNA"]
plot_category_chart(df=app_train_gender, feature="CODE_GENDER", x_name=["Male", "Female"], rotation=0)
```

```python
plot_category_chart(df=app_train, feature="NAME_INCOME_TYPE", rotation=45)
```

```python
plot_category_chart(df=app_train, feature="ORGANIZATION_TYPE", rotation=90, ascending=True, figsize=(20, 8))
```

```python
plot_category_chart(df=app_train_gender, feature="NAME_EDUCATION_TYPE", rotation=45)
```
