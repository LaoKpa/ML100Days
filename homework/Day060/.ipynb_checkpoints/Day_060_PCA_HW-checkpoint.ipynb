# 使用手寫辨識資料集, 觀察 PCA 算法

# [作業目標]
- 試著模仿範例寫法, 以變更參數後的 PCA + 邏輯斯迴歸判斷手寫辨識資料集, 觀察不同 component 下正確率的變化

# [作業重點]
- 以手寫辨識資料集, 觀察變更參數後的 PCA 算法, 在取不同 component 時, PCA 解釋度與分類正確率如何變化 (In[5], Out[5])

```python
# 載入套件
import multiprocessing
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
import warnings
warnings.filterwarnings("ignore")
```

```python
# 定義 PCA 與隨後的羅吉斯迴歸函數
# penalty 改為 L1, max_iter 改為 100
logistic = SGDClassifier(loss='log', penalty='l1', max_iter=100, tol=1e-5, random_state=0)
pca = PCA()
pipe = Pipeline(steps=[('pca', pca), ('logistic', logistic)])

# 載入手寫數字辨識集
digits = datasets.load_digits()
X_digits = digits.data
y_digits = digits.target
```

# 作業
* 將參數 penalty 改為 L1, max_iter 改為 100, 觀察 PCA 與羅吉斯迴歸做手寫辨識的效果

```python
# 先執行 GridSearchCV 跑出最佳參數
n_components = [4, 10, 20, 30, 40, 50, 64]
penalty = ["l1", "l2"]
warm_start = [True, False]
shuffle = [True, False]
param_grid = dict(
    pca__n_components=n_components, 
    logistic__penalty=penalty, 
    logistic__warm_start=warm_start, 
    logistic__shuffle=shuffle)

# Random search
grid_search = GridSearchCV(
    pipe, 
    param_grid, 
    scoring="accuracy", 
    cv=ShuffleSplit(n_splits=10, test_size=0.3, random_state=417), 
    n_jobs=multiprocessing.cpu_count(), 
    verbose=1)
rand_result = grid_search.fit(X_digits, y_digits)
print(rand_result.best_params_)
```

```python
# 繪製不同 components 的 PCA 解釋度
plt.style.use("seaborn")
pca.fit(X_digits)

fig, (ax0, ax1) = plt.subplots(nrows=2, sharex=True, figsize=(15, 10))
ax0.plot(pca.explained_variance_ratio_, linewidth=2)
ax0.set_ylabel('PCA explained variance')

ax0.axvline(grid_search.best_estimator_.named_steps['pca'].n_components, linestyle=':', label='n_components chosen')
ax0.legend(prop=dict(size=12))

# 繪製不同採樣點的分類正確率
results = pd.DataFrame(grid_search.cv_results_)
components_col = 'param_pca__n_components'
best_clfs = results.groupby(components_col).apply(lambda g: g.nlargest(1, 'mean_test_score'))

best_clfs.plot(x=components_col, y='mean_test_score', yerr='std_test_score', legend=False, ax=ax1)
ax1.set_ylabel('Classification accuracy (val)')
ax1.set_xlabel('n_components')
plt.tight_layout()
plt.show()
```
