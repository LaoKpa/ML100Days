# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def relu(x):
    return abs(x) * (x > 0)

def drelu(x):
    return (1 * (x > 0))
```

```python
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(15, 4))
plt.plot(x, relu(x), 'b')
plt.grid()
plt.title('ReLU Function')
plt.ylabel(None)
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(15, 4))
plt.plot(x, drelu(x), 'b')
plt.grid()
plt.title('Derivative ReLU Function')
plt.xlim((-10, 10))
plt.show()
```
