# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
def relu(x):
    return np.maximum(0.0, x)

def drelu(x):
    x[x <= 0] = 0.0
    x[x > 0] = 1.0
    return x
```

```python
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(15, 4))
plt.plot(x, relu(x), 'b')
plt.grid()
plt.title('ReLU Function')
plt.ylabel(None)
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(15, 4))
plt.plot(x, drelu(x), 'b')
plt.grid()
plt.title('Derivative ReLU Function')
plt.xlim((-10, 10))
plt.show()
```
